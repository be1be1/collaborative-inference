```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("GoogLeNet.xls")
```

```python
df
```

```python
tasks = list(df['task'])
```

```python
temp = []
```

```python
tasks
```

```python
for i, v in enumerate(list(df['c_task'])):
    temp.append(tasks[i])
    if type(v) == float:
        continue
    elif "," in v:
        temp.extend(v.split(","))
    else:
        temp.append(v)
        
```

```python
temp
```

```python
name_map = {}
```

```python
for i, v in enumerate(temp):
    name_map[v] = i
```

```python
name_map
```

```python
df['task'] = df['task'].map(name_map)
```

```python
df
```

```python
def mapper_c_task(x):
    if type(x) == float:
        return
    elif "," in x:
        return ",".join([str(name_map[i.strip()]) for i in x.split(",")])
    else:
        return name_map[x]
```

```python
df['c_task'] = df['c_task'].apply(mapper_c_task)
df['dir_succ'] = df['dir_succ'].apply(mapper_c_task)
df['dir_succ_add'] = df['dir_succ_add'].apply(mapper_c_task)
```

```python
df
```

```python
df = df.loc[:,df.columns[:-1]]
```

```python
df
```

```python
df.info()
```

```python
def get_count(x):
    if type(x)!=str:
        return 1
    elif type(x) == float:
        return 
    else:
        return len(x.split(","))
```

```python
df['count'] = df['c_task'].apply(get_count)
```

```python
df
```

```python
def transfer(x):
    tl = x.split("|")
    if tl[1] == 1:
        return tl[0]
    else:
        res = []
        for i in range(int(tl[1])):
            res.append(str(np.round(float(tl[0]),5)))
        return ",".join(res)
```

```python
df['count'] = df['count'].apply(str)
```

```python
for col in df.columns[8:24]:
    df[col] = df[col].apply(str)
    df[col] = df[col] + "|" + df['count']
    df[col] = df[col].apply(transfer)
```
